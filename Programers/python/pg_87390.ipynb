```python
'''
가장 접근 하기 쉬운 방법으로 풀었으나 역시 시간 초과
'''
n, left, right = 5	,2	,5

def solution(n, left, right):
    answer = []
    
    arr = [[0 for _ in range(n)] for _ in range(n)]
    
    for i in range(n):
        c = i + 1
        arr[i][i] = c
        for j in range(c):
            arr[j][i] = c
            arr[i][j] = c
    
    answer = [b for a in arr for b in a]
    
    return answer[left:right+1]


solution(n, left, right)
```

```python
'''
실패 
70점 
'''
def solution(n, left, right):
    answer = []
    arr = []
    nIdx, startIdx = left//n, left%n
    
    if startIdx - nIdx > 0 : 
        arr = [startIdx + 1]
    elif startIdx - nIdx < 0 : 
        arr = [nIdx+1] * (abs(startIdx-nIdx)+1)
    else :
        arr = [nIdx+1]
    
    runIdx = right - left
    if len(arr) > runIdx:
        return arr[:runIdx+1]
    
    rowNum = nIdx + 1
    while runIdx > 0 :
        if arr[-1] == n :
            rowNum += 1
            if rowNum == n :
                arr += [rowNum] * runIdx
                break
            
            if rowNum > runIdx :
                arr += [rowNum] * runIdx
                break
            
            arr += [rowNum] * rowNum
            runIdx -= rowNum
            
        else :
            arr.append(arr[-1] + 1)
            runIdx -= 1

    return arr
```

```python
'''
left, right사이를 for문을 돌려 해당 index의 들어갈 숫자를 구하는 식으로 했다.
i // n 은 점점 작아지는 특징, i % n 은 점점 커지는 특징을 가지고 있기 때문에
아래와 같은 식이 성립이 가능하다.
'''
def solution(n, left, right):
    answer = []
    
    for i in range(left, right+1):
        t = i // n - i % n
        t = t if t > 0 else 0
        answer.append(i % n + 1 + t)
    
    return answer
```
