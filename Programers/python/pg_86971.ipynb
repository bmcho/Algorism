```python
'''
풀이시간을 최대한 줄여보려고했다가 실패했다.

가설 :가장 노드가 많이 연결 된 송전탑들 중 하나를 끊어야지만 끊어진 송전탑들의 개수가 비슷해진다.

위 가설로 가장 노드가 많은 송전탑에 연결 된 송전탑들의 개수를 구하고 차의 절대값 중 가장 작은것을 구하면 된다.

반례 : 송전탑들이 전부 같으 노드의 갯수로 연결이 되어있다면.. 통과 x
'''

def solution(n, wires):

  def getWiresCount(wire, dCopy):

    q = [] 
    for i in dCopy[wire]:
      q.append(i)
    
    visited = [0] * (n + 1)
    visited[wire] = 1

    cnt = 1
    while q:
      z = q.pop(0)
      visited[z] = 1
      cnt += 1
      for i in dCopy[z] :
        if visited[i] : continue
        else : q.append(i)

    return cnt

  answer = 9999

  d = [[] for _ in range(n+1)]
  c = [0] * (n + 1)

  for wire in wires:

    v1,v2 = wire
    d[v1].append(v2)
    c[v1] += 1
    d[v2].append(v1)
    c[v2] += 1

  maxWiresIndex = c.index(max(c))
  maxWiresList = d[maxWiresIndex]

  for wire in maxWiresList:

    #각 송전탑을 7과 분리하고 연결된 갯수를 구한다. 가장 많이 연결된 송전탑의 갯수만 구하면 됨
    # deepcopy
    dCopy = [item[:] for item in d]
    dCopy[wire].remove(maxWiresIndex)
    dCopy[maxWiresIndex].remove(wire)

    if not dCopy[wire] : continue
    disConCnt1 = getWiresCount(wire, dCopy) + 1
    disConCnt2 = getWiresCount(maxWiresIndex, dCopy) + 1

    answer = min(answer, abs(disConCnt1-disConCnt2))
  
  return answer
```

```python
'''
모든 연결된 wire들을 하나씩 다 끊어서 bfs로 구함

collection 적극 활용하자 편하고 빠르다
'''

from collections import defaultdict, deque

def solution(n, wires):

  def bfs(s, v1,v2):
    
    q = deque()
    q.append(s)
    visited[s] = 1
    cnt = 1
    while q:
      z = q.popleft()

      for i in d[z] :
        if not ((z == v1 and i == v2) or (z == v2 and i == v1)) :
          if not visited[i] :
            visited[i] = 1
            q.append(i)
            cnt +=1

    return cnt

  answer = 9999
  d = defaultdict(list)

  for wire in wires:
    v1,v2 = wire
    d[v1].append(v2)
    d[v2].append(v1)

  for wire in wires:
    visited = [0] * (n + 1)
    v1, v2 = wire
    t = []

    for i in range(1, n+1):
      if not visited[i]:
        t.append(bfs(i, v1, v2))

    answer = min(answer, abs(t[0] - t[1]))
  
  return answer
```
