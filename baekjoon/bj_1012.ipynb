### [1012번 유기농배추](https://www.acmicpc.net/problem/1012)
### bfs 또는 dfs 로 풀면 되는 문제

```python
'''
bfs시간초과..
'''
from sys import stdin
from queue import Queue

result =[]

t = int( stdin.readline())
dr = [0,1,0,-1]
dc = [1,0,-1,0]

for _ in range(t):
    r,c,k = map(int, stdin.readline().split())
    visit = [[-1 for _ in range(c)] for _ in range(r)]
    napas = []
    q = Queue()

    for _ in range(k):
        n1, n2 = map(int, stdin.readline().split())
        napas.append([n1, n2])
    
    worm = 0
    for napa in napas :
        m,n = napa
        if visit[m][n] == -1 :
            q.put(napa)
            worm += 1
            while q.qsize() > 0 :
                m,n = q.get()
                visit[m][n] = 1

                for i in range(4) :
                    mr = m + dr[i]
                    mc = n + dc[i]

                    if mr < 0 or mc < 0 or mr >= r or mc >= c :
                        continue

                    if not [mr, mc] in napas or visit[mr][mc] == 1 :
                        continue

                    q.put([mr, mc])
    
    print(worm)
```

```python
'''
런타임 에러 (RecursionError)
python의 경우 백준에서는 dfs깊이를 1000으로 제한을 두고있다.
dfs의 경우 50*50 =2500이므로 최대깊이가 2500이다 
'''

from sys import stdin

dy = [0,1,0,-1]
dx = [1,0,-1,0]

def dfs(y, x, visit, napas, r, c):
    visit[y][x] = 1

    for i in range(4) :
        my = y + dy[i]
        mx = x + dx[i]

        if my < 0 or mx < 0 or my >= r or mx >= c :
            continue    

        if not [my, mx] in napas or visit[my][mx] == 1 :
            continue      

        dfs(my,mx,visit, napas, r, c)

t = int( stdin.readline())
for _ in range(t):
    r,c,k = map(int, stdin.readline().split())
    visit = [[-1 for _ in range(c)] for _ in range(r)]
    napas = []
    for _ in range(k):
        n1, n2 = map(int, stdin.readline().split())
        napas.append([n1, n2])
    
    worm = 0
    for y,x in napas :
        
        if visit[y][x] == -1 :
            worm += 1
            dfs(y,x, visit, napas, r, c)
    
    print(worm)
```

```python
'''
다시 dfs 성공
아마 위 두개의 코드는 잘못된 부분이 있었던것 같다, 또한 위 두개의 코드는 매트릭스를 파악하기 어렵게 해놓았다..
'''
from sys import stdin

dx = [1,-1,0,0]
dy = [0,0,1,-1]


t = int(stdin.readline())
for _ in range(t):
    m,n,k = map(int, stdin.readline().split())

    a = [[0] * m for _ in range(n)]
    v = [[0] * m for _ in range(n)]
    b = []

    for _ in range(k) :
        x,y = map(int, stdin.readline().split())
        a[y][x] = 1
        b.append([x,y])

    result = 0
    for x,y in b :
        if a[y][x] == 1 and v[y][x] == 0 :
            result += 1
            v[y][x] = 1
            q = [[x,y]]
            while q :
                x,y = q.pop(0)
                # print(x,y)
                for i in range(4) :
                    mx = x + dx[i]
                    my = y + dy[i]
                    
                    if 0 <= mx < m and 0 <= my < n and v[my][mx] == 0 and a[my][mx] == 1:
                        v[my][mx] = 1
                        q.append([mx,my])

    print(result)
```
