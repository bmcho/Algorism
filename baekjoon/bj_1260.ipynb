### [1260 DFS와BFS](https://www.acmicpc.net/problem/1260)

```python
'''
실패
시작지점의 간선이 없을 경우 에러가 난다.
'''
from sys import stdin

def dfs(v):
  result_dfs.append(v)
  visited_dfs[v] = 1

  for i in course[v]:
    if not visited_dfs[i]:
      dfs(i)

def bfs(v):
  result_bfs.append(v)
  visited_bfs[v] = 1

  q = []
  q.extend(course[v])

  while q :
    c = q.pop(0)

    if not visited_bfs[c] :
      visited_bfs[c] = 1
      result_bfs.append(c)
      q.extend(course[c])


N,M,V = map(int, stdin.readline().split())

visited_dfs = [0] * (N+1)
visited_bfs = [0] * (N+1)

course = dict()

for _ in range(M):
  a,b = map(int,stdin.readline().split())

  if a in course.keys() :
    course[a].append(b)
  else :
    course[a] = [b]

  if b in course.keys() :
    course[b].append(a)
  else :
    course[b] = [a]

for key, item in course.items():
  item.sort()

result_dfs = []
result_bfs = []

dfs(V)
bfs(V)

print(' '.join(map(str,result_dfs)))
print(' '.join(map(str,result_bfs)))
```

```python
'''
시작지점의 간선이 없을경우
따로 빈배열을 추가 해주었다.
'''

from sys import stdin

def dfs(v):
  result_dfs.append(v)
  visited_dfs[v] = 1

  for i in course[v]:
    if not visited_dfs[i]:
      dfs(i)

def bfs(v):
  result_bfs.append(v)
  visited_bfs[v] = 1

  q = []
  q.extend(course[v])

  while q :
    c = q.pop(0)

    if not visited_bfs[c] :
      visited_bfs[c] = 1
      result_bfs.append(c)
      q.extend(course[c])


N,M,V = map(int, stdin.readline().split())

visited_dfs = [0] * (N+1)
visited_bfs = [0] * (N+1)

course = dict()

for _ in range(M):
  a,b = map(int,stdin.readline().split())

  if a in course.keys() :
    course[a].append(b)
  else :
    course[a] = [b]

  if b in course.keys() :
    course[b].append(a)
  else :
    course[b] = [a]

for key, item in course.items():
  item.sort()

'''
추가된 부분
'''
if V not in course:
  course[V] = []

result_dfs = []
result_bfs = []

dfs(V)
bfs(V)

print(' '.join(map(str,result_dfs)))
print(' '.join(map(str,result_bfs)))
```
