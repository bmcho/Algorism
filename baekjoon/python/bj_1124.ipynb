### [1124번 언더프라임](https://www.acmicpc.net/problem/1124)

### 자연수 X의 소인수 분해 로직과 소수구하는 로직을 사용하면 가능

```python
'''
시간초과
'''
A, B = map(int,input().split())

def isPrime(num) :
    if num == 1 : return False
    n = int(num ** 0.5) + 1
    for i in range(2,n) :
        if num % i == 0 :
            return False
    
    return True

def countFactorization(num) :
    n = 2
    r = 0
    while num != 1 :
        if num % n == 0 :
            num = num//n
            r += 1
        else :
            n += 1

    return r

# print(countFactorization(456))

result = 0
for i in range(A, B + 1) :
    n = countFactorization(i)
    if isPrime(n) :
        result += 1

print(result)
```

```python
'''
시간초과
소인수분해쪽을 더 줄여봐야겠다
'''
A, B = map(int,input().split())

def countFactorization(num) :
    n = 2
    r = 0
    while num != 1 :
        if num % n == 0 :
            num = num//n
            r += 1
        else :
            n += 1

    return r

prime = [True] * (B + 1)
prime[0] = False
prime[1] = False

for i in range(2,int(B**0.5)+1) :
    if prime[i] == True :
        for j in range(i+i, B + 1, i) :
            prime[j] = False

result = 0
for i in range(A, B + 1) :
    if prime[i] : continue
    n = countFactorization(i)
    if prime[n] :
        result += 1

print(result)
```

```python
'''
성공

소인수분해를 dp 로 변경하여 품

4의 경우 2 * 2 로 구성되어있으며 2의 인수개수 + 1를 하면 4의 인수 개수가 됨
8의 경우 4의 인수 개수 + 1 > 3
'''
A, B = map(int,input().split())

prime = [True] * (B + 1)
dp = [1] * (B + 1)
prime[0] = False
prime[1] = False

for i in range(2,int(B**0.5)+1) :
    if prime[i] == True :
        for j in range(i+i, B + 1, i) :
            prime[j] = False

for i in range(2,B+1) :
    for j in range(2,B+1):
        if i * j > B :
            break
        dp[i*j] = dp[i] +1

result = 0
for i in range(A, B + 1) :
    if prime[i] : continue
    if prime[dp[i]] :
        result += 1

print(result)
```

```python
'''
다른사람의 풀이
소수, 인수개수 두개의 list를 두지 않고 하나의 list로만 처리
소수는 인수개수가 하나이므로 따로 구분을 하지 않아도 인수개수가 하나면 소수가 되는 것!!


factor의 배수만큼 num_of_factors[i]+=1 해주어 인수개수를 계산
16의 경우 16부터 2씩 나눈다고 생각하면 +=1 하는 이유를 조금 이해할수 있음
'''

a, b = map(int, input().split())
num_of_factors = [0] * (b + 1)

for n in range(2, b + 1):
    if num_of_factors[n] == 0:
        factor = n
        while factor <= b:
            for i in range(factor, b + 1, factor):
                num_of_factors[i] += 1
            factor *= n
            
num_of_underprimes = 0
for count in num_of_factors[a:]:
    if num_of_factors[count] == 1:
        num_of_underprimes += 1
        
print(num_of_underprimes)
```
