[파이프 옮기기1](https://www.acmicpc.net/problem/17070)

```python
'''
bfs 시간초과
'''
import sys
from collections import deque

n = int(sys.stdin.readline())

grid = []

for _ in range(n):
    grid.append(list(map(int,sys.stdin.readline().split())))

q = deque()

q.append([(0,0),(0,1)])

answer = 0
while q:
    f,l = q.popleft()

    if l[0] == n-1 and l[1] == n-1:
        answer += 1
        continue

    #가로
    if l[0] - f[0] == 0 and l[1] - f[1] == 1:
        if l[1] + 1 < n and grid[l[0]][l[1]+1] != 1:
            q.append([l,(l[0],l[1]+1)])
    #세로
    elif l[0] - f[0] == 1 and l[1] - f[1] == 0:
        if l[0] + 1 < n and grid[l[0]+1][l[1]] != 1:
            q.append([l,(l[0]+1,l[1])])
    #대각
    elif l[0] - f[0] == 1 and l[1] - f[1] == 1:
        if l[1] + 1 < n and grid[l[0]][l[1]+1] != 1:
            q.append([l,(l[0],l[1]+1)])
        if l[0] + 1 < n and grid[l[0]+1][l[1]] != 1:
            q.append([l,(l[0]+1,l[1])])

    #대각 공통
    if l[0] + 1 < n and l[1] + 1 < n:
        if grid[l[0]+1][l[1]] != 1 and grid[l[0]][l[1]+1] != 1 and grid[l[0]+1][l[1]+1] != 1:
            q.append([l,(l[0]+1,l[1]+1)])   

print(answer)
```

```python
'''
PyPy dfs 성공
'''
import sys

n = int(sys.stdin.readline())

grid = []

for _ in range(n):
    grid.append(list(map(int,sys.stdin.readline().split())))

answer = 0

def dfs(p):
    x,y,z = p
    global answer
    if x == n-1 and y == n-1:
        answer += 1
        return
    
    #대각
    if x + 1 < n and y + 1 < n :
        if grid[x+1][y] != 1 and grid[x][y+1] != 1 and grid[x+1][y+1] != 1:
            dfs((x+1,y+1,2))
    
    #가로
    if z == 0 or z == 2:
        if y + 1 < n and grid[x][y+1] != 1:
            dfs((x,y+1,0))

    if z == 1 or z == 2:
        if x + 1 < n and grid[x+1][y] != 1:
            dfs((x+1,y,1))

dfs((0,1,0))

print(answer)
```

```python
'''
dp 성공
'''
import sys

n = int(sys.stdin.readline())

grid = []

for _ in range(n):
    grid.append(list(map(int,sys.stdin.readline().split())))

dp = [[[0] * n for _ in range(n)] for _ in range(3)]


dp[0][0][1] = 1

for i in range(2,n):
    if grid[0][i] == 1 : break
    dp[0][0][i] = 1


for r in range(1,n):
    for c in range(1,n):
        #대각
        if grid[r][c] != 1 and grid[r][c-1] != 1 and grid[r-1][c] != 1:
            dp[2][r][c] = dp[0][r-1][c-1] + dp[1][r-1][c-1] + dp[2][r-1][c-1]

        if grid[r][c] == 0:
            # 현재 위치가 가로인 경우
            dp[0][r][c] = dp[0][r][c - 1] + dp[2][r][c - 1]
            # 현재 위치가 세로인 경우
            dp[1][r][c] = dp[1][r - 1][c] + dp[2][r - 1][c]


print(sum(dp[i][-1][-1] for i in range(3)))
```
